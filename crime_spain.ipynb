```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv('Datos/crime_rate_Spain.csv',sep=',',header = 0)
data.info()
```

```python
# Filtrar los datos para obtener solo las filas con 'Location' igual a 'Barcelona'
data_barcelona = data[data['Location'] == 'Barcelona']

# Crear un gráfico de dispersión con Seaborn
sns.scatterplot(data=data_barcelona, x='Year', y='Total cases', hue='Year')

# Ajustar el Total de casos (y) con una escala logarítmica si es necesario
plt.yscale('log')

# Agregar leyenda y título
plt.legend(title='Year')
plt.xlabel('Year')
plt.ylabel('Total cases')
plt.title('Total cases in Barcelona over the years')

# Mostrar el gráfico
plt.show()
```

```python
# Filtrar los datos para obtener solo las filas con 'Location' igual a 'Seville'
data_seville = data[data['Location'] == 'Seville']

# Crear un gráfico de barras con Seaborn
sns.barplot(data=data_seville, x='Year', y='Total cases')

# Agregar etiquetas y título
plt.xlabel('Year')
plt.ylabel('Total cases')
plt.title('Total cases in Seville over the years')

# Mostrar el gráfico
plt.show()
```

```python
# Filtrar los datos para obtener solo las filas con 'Location' igual a 'Murcia'
data_city = data[data['Location'] == 'Murcia']

# Crear un gráfico de barras con Seaborn
sns.boxplot(data=data_city, x='Year', y='Total cases')

# Agregar etiquetas y título
plt.xlabel('Year')
plt.ylabel('Total cases')
plt.title('Total cases in Murcia over the years')

# Mostrar el gráfico
plt.show()
```
